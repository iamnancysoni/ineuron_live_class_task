# 9 . Average weekly response time for each agent

```python
df2.columns
```

```python
df2['Average_response_time'] = pd.to_datetime(df2['Average Response Time'])
```

```python
df2['hrs']=df2['Average_response_time'].dt.hour
df2['mins']=df2['Average_response_time'].dt.minute
df2['sec']=df2['Average_response_time'].dt.second
df2['mins']=df2['mins']*60
df2['time in sec']=df2['mins']+df2['sec']
```

```python
df2.head()
```

```python
df2.groupby(['Agent Name','week'])['time in sec'].mean().to_frame()
```

# 10 . Average weekely resolution time for each agents

```python
df2['Average_Resolution_Time'] = pd.to_datetime(df2['Average Resolution Time'])
```

```python
df2['hrs_art']=df2['Average_Resolution_Time'].dt.hour
df2['mins_art']=df2['Average_Resolution_Time'].dt.minute
df2['sec_art']=df2['Average_Resolution_Time'].dt.second
df2['hrs_art']=df2['hrs_art']*3600
df2['mins_art']=df2['mins_art']*60
```

```python
df2['art_time_in_sec']=df2['hrs_art']+df2['mins_art']+df2['sec_art']
```

```python
df2.groupby(['Agent Name','week'])['art_time_in_sec'].mean().to_frame()
```

# 13 . Total contributation hour for each and every agents weekly basis

```python
df1
```

```python
df1['updated_duration'] = (pd.to_datetime(df1.Date) + pd.to_timedelta(df1.Duration))
```

```python
df1['hrs_contribution']=df1['updated_duration'].dt.hour
df1['min_contribution']=df1['updated_duration'].dt.minute
df1['sec_contribution']=df1['updated_duration'].dt.second
df1['hrs_contribution']=df1['hrs_contribution']*3600
df1['min_contribution']=df1['min_contribution']*60
```

```python
df1['total_contribution_in_sec']=df1['hrs_contribution']+df1['min_contribution']+df1['sec_contribution']
```

```python
df1['total_contribution_in_hrs']=df1['total_contribution_in_sec']/3600
```

```python
df1['week']=df1['updated_duration'].dt.week
```

```python
df1
```

```python
df1.groupby(['Agent','week'])['total_contribution_in_hrs'].sum().to_frame()
```
